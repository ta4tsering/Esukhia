```python
from pathlib import Path
from bs4 import BeautifulSoup
import re 
```

```python
def poti_parse(content):
    result = ''
    temp= ''
    num = 1
    pages = re.split(r'\n\n', content)
    for pg_num, page in enumerate(pages, 1):
        if re.search('\[I1PD(.+)\]', page):
            pg_id = re.search('\[I1PD(.+)\]', page).group(1)
            if pg_num % 2 != 0:
                new_pg_content = re.sub('I1PD.+', f'{num}a] ', page) 
            elif pg_num%2 == 0:
                new_pg_content = re.sub('I1PD.+', f'{num}b] ', page)
                num += 1  
        else:
            if pg_num % 2 != 0:
                new_pg_content = f'[{num}a]\n{page}'
            elif pg_num%2 == 0: 
                new_pg_content = f'[{num}b]\n{page}'
                num += 1
        result += new_pg_content + '\n\n'
    return result
```

```python
def poti_update(content,poti_num):
    phudrak_content = Path(f'../numbered_correctly_tengyur_namsel/v{poti_num:03}').write_text(content, encoding='utf-8')
    print("Update comoplete .....")
```

```python
def get_poti_content(output_path, poti_num):
    page_content = output_path.read_text(encoding='utf-8')
    poti_content = poti_parse(page_content)
    poti_update(poti_content, poti_num)
```

```python
def get_potis():
    all_potis = list(Path(f'../tengyur_namsel_text').iterdir())
    all_potis.sort()
    
    for poti_num, poti_path in enumerate(all_potis[2:],1):
        poti_id = poti_path.name
        print(f'{poti_path.stem} processing....')
        get_poti_content(poti_path, poti_num)
```
