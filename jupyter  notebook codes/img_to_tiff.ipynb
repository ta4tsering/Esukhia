```python
from pathlib import Path
import cv2
```

```python
def update_photo(img_bw,photo_id):
    img_path = Path(f'../../namsel_output/{photo_id}.tif')
    loc = str(img_path)
    cv2.imwrite(loc, img_bw)
    print(f'update complete for {photo_id} done.')
```

```python
def to_tiff(photo_path):
    img_gray = cv2.imread(str(photo_path), cv2.IMREAD_GRAYSCALE)
    print(f'{img_gray.shape}')
    (thresh, img_bw) = cv2.threshold(img_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    return img_bw
```

```python
def get_photos():
    folder_path = list(Path(f'../../namsel').iterdir())
    folder_path.sort()
    for photo_num, photo_path in enumerate(folder_path[1:],1):
        photo_id = photo_path.stem
        print(f'Photo {photo_id} processing ...')
        img_bw = to_tiff(photo_path)
        update_photo(img_bw, photo_id)
        
```

```python
get_photos()
```
