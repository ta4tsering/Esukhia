```python
def update_repo(g, pecha_id, file_path, commit_msg, new_content):
    try:
        repo = g.get_repo(f"Openpecha/{pecha_id}")
        contents = repo.get_contents(f"{file_path}", ref="master")
        repo.update_file(contents.path, commit_msg , new_content, contents.sha, branch="master")
        print(f'{pecha_id} update completed..')
    except:
        print('Repo not found')
```

```python
def get_readme(g, pecha_id):
    try:
        repo = g.get_repo(f"Openpecha/{pecha_id}")
        contents = repo.get_contents("README.md")
        return contents.decoded_content.decode()
    except:
        print('Repo Not Found')
        return ''
```

```python
def get_updated_readme(old_readme, pecha_id):
    editor_link = f'[<img src="https://img.icons8.com/color/25/000000/edit-property.png"> Open in Editor](http://editor.openpecha.org/{pecha_id})'
    bdrc_link = f'[<img width="25" src="https://library.bdrc.io/icons/BUDA-small.svg"> Open in Buda](https://library.bdrc.io/show/bdr:IE0OP{pecha_id})'
    new_readme = f"{old_readme}\n\n{editor_link}\n\n{bdrc_link}\n"
    return new_readme
```

```python
with open("catalog.csv", newline="") as csvfile:
    pechas = list(csv.reader(csvfile, delimiter=","))
    for pecha in pechas[278:]:
        pecha_id = re.search("\[.+\]", pecha[0])[0][1:-1]
        old_readme = get_readme(g, pecha_id)
        if "Open in Buda" in old_readme:
            pass
        else:
            new_content = get_updated_readme(old_readme, pecha_id)
            update_repo(g, pecha_id, file_path, commit_msg, new_content)
```
