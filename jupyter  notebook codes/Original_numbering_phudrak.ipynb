```python
from pathlib import Path
import re 
```

```python
def poti_parse(content):
    result = ''
    pages = re.split(r'\n\n', content)
    for pg_num, page in enumerate(pages, 1):
        if re.search('Page', page):
            pg_id = re.search('Page no:(.+)', page).group(1)
            new_pg_content = re.sub('Page no:.+', f'[{pg_num}]{pg_id}', page) 
        else:
            new_pg_content = f'[{pg_num}]\n{page}'
        result += new_pg_content + '\n\n'
    return result
```

```python
def poti_update(content,poti_num):
    phudrak_content = Path(f'./numbered_phudrak/v{poti_num:03}').write_text(content, encoding='utf-8')
    print("Update comoplete .....")
    
```

```python
def get_poti_content(output_path, poti_num):
    page_content = output_path.read_text(encoding='utf-8')
    poti_content = poti_parse(page_content)
    poti_update(poti_content, poti_num)
```

```python
def get_potis():
    all_potis = list(Path(f'./transcript').iterdir())
    all_potis.sort()
    
    for poti_num, poti_path in enumerate(all_potis[1:],1):
        poti_id = poti_path.name
        print(f'{poti_id} processing \n')
        get_poti_content(poti_path, poti_num)
```
