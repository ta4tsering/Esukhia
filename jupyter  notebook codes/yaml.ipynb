```python
from pathlib import Path
import re
import yaml
```

```python
def adding_durchen_annotation(content,start,end): 
    result = ''
    pages = get_pages(content)
    print("Poti parse works")
    for pg_num, page in enumerate(pages, 1):
        if start-1 == pg_num:
            result += f'<d{page}'
        elif end-1 == pg_num:
                result += f'{page}/d>' 
        else:
            result += page
    print("end of poti parse")
    return result
```

```python
def get_pages(vol_text):
    result = []
    pg_text = ""
    pages = re.split(f"(\[\d+\w+\])", vol_text)
    for i, page in enumerate(pages[1:]):
        if i % 2 == 0:
            pg_text += page
        else:
            pg_text += page
            result.append(pg_text)
            pg_text = ""
    return result
```

```python
def poti_update(content,poti_num):
    namsel_content = Path(f'./new_google/v{poti_num:03}').write_text(content, encoding='utf-8')
    print(f'V{poti_num:03} Update complete .....')
```

```python
def get_location(page_content, poti_num, t2d):
    for text_id, text in t2d.items():
        pg_start = int(text['durchen']['img_loc_start'])
        pg_end = int(text['durchen']['img_loc_end'])
        poti_content = adding_durchen_annotation(page_content, pg_start, pg_end)
        print(pg_start, pg_end)
        print("this is it ")
        poti_update(poti_content,poti_num)
        print("poti update works")
```

```python
def get_poti_content(output_path, poti_num, t2d):
    page_content = output_path.read_text(encoding='utf-8')
    get_location(page_content, poti_num, t2d)
    
```

```python
def get_potis():
    text_2_durchen = Path(f'./kangyur_t2d.yaml').read_text(encoding = 'utf-8')
    t2d = yaml.safe_load(text_2_durchen)
    all_potis = list(Path(f'./namsel-ocr/clean_google_text').iterdir())
    all_potis.sort()
    for poti_num, poti_path in enumerate(all_potis,1):
        poti_id = poti_path.name
        print(f'{poti_id} processing \n')
        get_poti_content(poti_path, poti_num, t2d)
```
